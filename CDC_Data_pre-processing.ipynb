# Import primary CDC data

```python
data = Path('data/PLACES__County.csv')
county_df = pd.read_csv(data)
county_df.head(5)
```

## Data pre-processing 

```python
#drop confidence interval columns
county_pop_df = county_df.filter(['StateDesc', 'CountyName', 'CountyFIPS', 'TotalPopulation', 'ACCESS2_CrudePrev'
                                  'ACCESS2_AdjPrev', 'ARTHRITIS_CrudePrev', 'ARTHRITIS_AdjPrev', 'BINGE_CrudePrev',
                                  'BINGE_AdjPrev', 'BPHIGH_CrudePrev', 'BPHIGH_AdjPrev', 'BPMED_CrudePrev', 'BPMED_AdjPrev',
                                  'CANCER_CrudePrev', 'CANCER_AdjPrev', 'CASTHMA_CrudePrev', 'CASTHMA_AdjPrev', 'CERVICAL_CrudePrev',
                                  'CERVICAL_AdjPrev',  'CHD_CrudePrev', 'CHD_AdjPrev', 'CHECKUP_CrudePrev', 'CHECKUP_AdjPrev', 
                                  'CHOLSCREEN_CrudePrev', 'CHOLSCREEN_AdjPrev', 'COLON_SCREEN_CrudePrev', 'COLON_SCREEN_AdjPrev',
                                  'COPD_CrudePrev','COPD_AdjPrev', 'COREM_CrudePrev', 'COREM_AdjPrev', 'COREW_CrudePrev', 'COREW_AdjPrev',
                                  'CSMOKING_CrudePrev', 'CSMOKING_AdjPrev', 'DENTAL_CrudePrev', 'DENTAL_AdjPrev', 'DEPRESSION_CrudePrev',
                                  'DEPRESSION_AdjPrev', 'DIABETES_CrudePrev', 'DIABETES_AdjPrev', 'GHLTH_CrudePrev', 'GHLTH_AdjPrev',
                                  'HIGHCHOL_CrudePrev', 'HIGHCHOL_AdjPrev', 'KIDNEY_CrudePrev', 'KIDNEY_AdjPrev', 'LPA_CrudePrev', 
                                  'LPA_AdjPrev', 'MAMMOUSE_CrudePrev', 'MAMMOUSE_AdjPrev', 'MHLTH_CrudePrev', 'MHLTH_AdjPrev', 
                                  'OBESITY_CrudePrev', 'OBESITY_AdjPrev', 'PHLTH_CrudePrev', 'PHLTH_AdjPrev', 'SLEEP_CrudePrev', 
                                  'SLEEP_AdjPrev', 'STROKE_CrudePrev', 'STROKE_AdjPrev', 'TEETHLOST_CrudePrev', 'TEETHLOST_AdjPrev',
                                  'Geolocation'], axis = 1)
county_pop_df.head(5)
```

```python
county_pop_df = county_pop_df.dropna()
```

```python
# Create for-loop to iterate through 'total pop' ccolumn and define as 'Rural' or 'Urban'
rural_urban = []
for population in county_pop_df['TotalPopulation']:
    if population < 50000:
        rural_urban.append("Rural")
    if population >= 50000:
        rural_urban.append("Urban")

county_pop_df['RuralUrban'] = rural_urban
```

# Import census county data for alternative Rural vs Urban classifier 

```python
data = Path('data/census_county_pop.csv')
census_df = pd.read_csv(data, skiprows=[1])
census_df.head(5)
```

## Data pre-processing

```python
#drop empty columns interval columns
census_df = census_df.filter(['FIPS', 'Name of Area', 'Qualifying Name', 'Population Density',
                              'Total Population', 'Population, Urban', 'Population, Rural'], axis = 1)
census_df.head(5)
```

```python
## . change name to remove spaces
```

```python
#create a new column the classifies countries as Rural or Urban based on population density
#create a list of conditions
conditions = [(census_df['Population Density']<=500),(census_df['Population Density'] >500)]

#create the list of values to assign for each condition
values = ['Rural', 'Urban']

#create new column and use np.select to assign values
census_df['ruralUrban'] = np.select(conditions, values)

census_df.head(5)
```
